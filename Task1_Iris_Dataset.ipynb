```python
# pandas: for loading and handling data
import pandas as pd

# seaborn: for beautiful statistical plots
import seaborn as sns

# matplotlib: for creating and customizing plots
import matplotlib.pyplot as plt

# warnings: to hide unnecessary warning messages
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully!")
```

```python
# Load the Iris dataset using seaborn
df = sns.load_dataset('iris')

# Show the first 5 rows to understand the structure
print("First 5 rows of the dataset:")
df.head()
```

```python
# Shape tells us rows and columns
print("Dataset Shape (rows, columns):", df.shape)

# Column names
print("\nColumn Names:", df.columns.tolist())

# Species (target classes)
print("\nSpecies (Target Classes):", df['species'].unique())

# Count of each species
print("\nCount of each species:")
print(df['species'].value_counts())
```

```python
# .info() shows data types and non-null counts
print("Dataset Info:")
df.info()
```

```python
# .describe() gives statistical summary (mean, min, max, std, etc.)
print("Statistical Summary:")
df.describe()
```

```python
# Check for missing values
print("Missing Values in each column:")
print(df.isnull().sum())
```

```python
# ---- SCATTER PLOT ----
# Shows relationship between Petal Length and Petal Width for each species

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',       # color by species
    palette='Set1',
    s=80                 # dot size
)
plt.title('Scatter Plot: Petal Length vs Petal Width', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# ---- HISTOGRAMS ----
# Shows the distribution (spread) of all 4 features

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Histograms: Distribution of All Features', fontsize=14, fontweight='bold')

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['steelblue', 'coral', 'mediumseagreen', 'mediumpurple']

for i, (feature, color) in enumerate(zip(features, colors)):
    row, col = i // 2, i % 2
    axes[row][col].hist(df[feature], bins=20, color=color, edgecolor='white', alpha=0.85)
    axes[row][col].set_title(feature.replace('_', ' ').title())
    axes[row][col].set_xlabel('Value (cm)')
    axes[row][col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# ---- BOX PLOTS ----
# Shows the median, spread, and outliers for each feature per species

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Box Plots: Feature Distribution by Species', fontsize=14, fontweight='bold')

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    sns.boxplot(
        data=df,
        x='species',
        y=feature,
        palette='pastel',
        ax=axes[row][col]
    )
    axes[row][col].set_title(feature.replace('_', ' ').title())
    axes[row][col].set_xlabel('Species')
    axes[row][col].set_ylabel('Value (cm)')

plt.tight_layout()
plt.show()
```
